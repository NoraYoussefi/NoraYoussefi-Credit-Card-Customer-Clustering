<h3 style="color:blue;">1. Définir le K nécessaire pour avoir un modèle efficace selon la méthode de Elbow.
<h3>


```python
# On a choisit pour la visualisation justes les features : BALANCE_FREQUENCY et PURCHASES
import os
import seaborn as sns
```

Le choix des features : BALANCE_FREQUENCY et PURCHASES n'est pas arbitraire, mais pour visualiser les montants des clients mis à jour par rapport à leurs montants d'achat.

```python
X = df.iloc[:, [2, 3]].values  

#finding optimal number of clusters using the elbow method  
from sklearn.cluster import KMeans  
wcss_list= []  #Initializing the list for the values of WCSS  
  
#Using for loop for iterations from 1 to 10.  
for i in range(1, 11):  
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 42)  
    kmeans.fit(X)  
    wcss_list.append(kmeans.inertia_)  
plt.plot(range(1, 11), wcss_list)  
plt.title('The Elobw Method Graph')  
plt.xlabel('Number of clusters(k)')  
plt.ylabel('wcss_list')  
plt.show() 
```

<h3 style="color:blue;">2. Essayer de construire le modèle de clustering en utilisant sklearn ainsi algorithme Kmeans.<h3>


```python
# Fitting K-Means to the dataset (K = 4) d'apres le graph
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

<h3 style="color:blue;">3. Présenter les cluster obtenues dans un graphes en utilisant matplotlib.<h3>


```python
# Visualising the clusters
plt.figure(figsize=(15,7))
sns.scatterplot(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], color = 'yellow', label ='Cluster 1',s=50)
sns.scatterplot(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], color = 'blue', label = 'Cluster 2',s=50)
sns.scatterplot(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], color = 'green', label ='Cluster 3',s=50)
sns.scatterplot(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], color = 'grey', label = 'Cluster 4',s=50)
sns.scatterplot(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color= 'red',label = 'Centroids',s=300,marker=',')
plt.grid(False)
plt.title('Clusters of customers')
plt.xlabel('BALANCE_FREQUENCY')
plt.ylabel('PURCHASES ')
plt.legend()
plt.show()
```

<h3 style="color:blue;">4. Interprétation<h3>


on observe que tant que les montants d'achats des clients augmentent, tant que leurs comptes se modifient et font plus des mise à jour c-ad la frequence augmente.

<h2 style="color:green">fuzzy cmeans<h2>

```python
import skfuzzy as fuzz
import seaborn as sns
import numpy as np

%matplotlib inline

colors = ['b', 'orange', 'g', 'r', 'c', 'm', 'y', 'k', 'Brown', 'ForestGreen']
```

In Fuzzy clustering each element has a set of membership coefficients corresponding to the degree of being in a given cluster. Points close to the center of a cluster, may be in the cluster to a higher degree than points in the edge of a cluster. The degree, to which an element belongs to a given cluster, is a numerical value varying from 0 to 1.

```python
# Define three cluster centers
centers = [[1, 3],
           [2, 2],
           [3, 8]]

# Define three cluster sigmas in x and y, respectively
sigmas = [[0.3, 0.5],
          [0.5, 0.3],
          [0.5, 0.3]]

# Generate test data
np.random.seed(42)  # Set seed for reproducibility
xpts = np.zeros(1)
ypts = np.zeros(1)
labels = np.zeros(1)
for i, ((xmu, ymu), (xsigma, ysigma)) in enumerate(zip(centers, sigmas)):
    xpts = np.hstack((xpts, np.random.standard_normal(200) * xsigma + xmu))
    ypts = np.hstack((ypts, np.random.standard_normal(200) * ysigma + ymu))
    labels = np.hstack((labels, np.ones(200) * i))

# Visualize the test data
fig0, ax0 = plt.subplots()
for label in range(3):
    ax0.plot(xpts[labels == label], ypts[labels == label], '.')
ax0.set_title('Test data: 200 points.')
plt.show()
```

```python
data =df.values
data = data[:, 1:4]
xpts = data[:, 0]
ypts = data[:, 1]
fig1, axes1 = plt.subplots(3, 3, figsize=(10, 10))
#Pour empiler les points(x,y)
alldata = np.vstack((xpts, ypts))
#Initialisation du tableau des fuzzy 
clustersfpcs = []
for ncenters, ax in enumerate(axes1.reshape(-1), 2):
    cntr, u, u0, d, jm, p,\
    fpc = fuzz.cluster.cmeans(alldata, ncenters, 2, error=0.005, maxiter=1000, init=None)
    #On ajoute à chaque fois au tableau le fuzzy cluster qu'on a
    clustersfpcs.append(fpc)
    
    # Pour tracer les clusters
    cluster_membership = np.argmax(u, axis=0)
    for j in range(ncenters):
        ax.plot(xpts[cluster_membership == j],
        ypts[cluster_membership == j], '.', color=colors[j])
    # Pour marquer le centre de chaque cluster
    for pt in cntr:
        ax.plot(pt[0], pt[1], 'rs')
    ax.set_title('Centers = {0}; FPC = {1:.2f}'.format(ncenters, fpc), size=12)
    ax.axis('off')
# Pour ajuster l'emplacement des axes
fig1.tight_layout()
```

Le C-means est capable de nous donner plus de clusters, et il est plus détaillé, car on peut avoir des appartiennent avec avec un certain porcentage à ce cluster et avec un autre pourcentage à un autre cluster.

Dans le cas de 4 clusters par exemple, on constate que les résultats sont les memes pour les personnes qui ont des montants d'achats élevés, mais pour ceux qui ont un montant d'achat bas, on constate qu'il y a des personnes de tout genre, ceux qui modifient leurs comptes toujours et ceux qui ne changent leurs comptes.

```python
#Pour dessiner plusieurs tracés sur une seule figure
fig2, ax2 = plt.subplots()
ax2.plot(np.r_[2:11], clustersfpcs, color='r')
ax2.set_title("How Number of Clusters Change FPC?")
ax2.set_xlabel("Number of centers")
ax2.set_ylabel("Fuzzy partition coefficient")
plt.show()
```

<h3 style="color:blue;">interpretation <h3>
Selon ce graphe, on constate que le nombre idéal des clusters qu'on doit utiliser est 6 et pas 4 comme le cas du K-means.

<h3 style="color:blue;">comparaison des deux algorithmes:
<h3>


Dans le cas de K-means,un objet ne peut pas appartenir à deux clusters simultanément,ce qui explique la discrimination binaire entre les clusters mais en Fuzzy C-means, il est possible qu’un objet appartient à deux ou plusieurs clusters selon différents pourcentages, c'est à dire que les données sont liées à chaque groupe par le biais d’une fonction d’appartenance, ce qui représente le comportement floue de cet algorithme.
Elles pourraient converger vers le même résultat,mais probablement avec un nombre différent d’itérations.
